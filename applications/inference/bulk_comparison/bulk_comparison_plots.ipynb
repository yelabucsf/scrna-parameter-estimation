### Squair et al datasets

```python
df1 = utils.canogamez_validation()
df2 = utils.hagai_validation()
df = pd.concat([df1, df2], axis=0)
```

```python
plt.figure(figsize=(2, 2))
my_pal = {
    "memento": "cyan", 
    "edger_lrt": "pink", 
    "edger_qlft":"pink",
    "deseq2_wald":"pink",
    "deseq2_lrt":"pink",
    "t":"cyan",
    "MWU":"cyan"
}
sns.boxplot(y='name', x='auc', data=df, palette=my_pal)
sns.stripplot(y='name', x='auc', data=df, color='grey')
plt.title('Squair et al. comparisons')
```

### Lupus data subsets

Similar to Squair et al's analysis above but with data from Perez et al (https://www.science.org/doi/10.1126/science.abf1970)


```python
pal = [(64/255,224/255,208/255)]+sns.color_palette("Greys", 10)[4:]
```

```python
df = utils.lupus_bulk_validation()
```

```python
plt.figure(figsize=(2.5,2))
sns.boxplot(
    y='name',
    x='auc', 
    data=df, 
    order=[
        'memento',
        'edgeR',
        'DESeq2',
        'MWU',
        't-test',],
    fliersize=0,
    palette=pal)
sns.stripplot(
    y='name',
    x='auc', 
    data=df, 
    order=[
        'memento',
        'edgeR',
        'DESeq2',
        'MWU',
        't-test',],
    linewidth=1,
    palette=pal)
plt.xlabel('Concordance to \nbulk RNA-seq')
plt.ylabel('')
```
