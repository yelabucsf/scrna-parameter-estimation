### GSEA

```r
ct = 'CD8 T cells'
```

```r
gene_list = df[!is.na(df$ENTREZID) & (df$ct == ct),]$dv_diff
names(gene_list) = df[!is.na(df$ENTREZID) & (df$ct == ct),]$ENTREZID
gene_list = sort(gene_list, decreasing=TRUE)
```

```r
length(gene_list)
```

```r
gsea <- gseKEGG(
    gene_list,
    organism = "hsa",
    keyType = "kegg",
    exponent = 1,
    nPerm = 1000,
    minGSSize = 10,
    maxGSSize = 500,
    pvalueCutoff = 0.1,
    pAdjustMethod = "BH",
    verbose = TRUE,
    use_internal_data = FALSE,
    seed = FALSE,
    by = "fgsea"
    )
```

```r
barplot(gene_list)
```
